# Edge Detector using OpenCV

```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15, 15]
```

```python
def image_processor(raw_img):
    img = cv2.cvtColor(raw_img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edge = cv2.Canny(gray, 100, 200)
    return img, gray, edge
```

```python
def display(dispImg, gray, edge):
    images = [dispImg, edge]
    titles = ['image', 'egde']

    for i in range(2):
        plt.subplot(2, 1, i+1)
        plt.imshow(images[i], 'gray')
        plt.title(titles[i])
        plt.show()
        
    cv2.imshow('edge', edge)
    cv2.imshow('real', dispImg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def save_image(new_edge):
    cv2.imwrite('edge-detect-result.png', new_edge)
```

```python
def edge_detector(query_image):
    new_img, new_gray, new_edge = image_processor(query_image)
    save_image(new_edge)
    display(new_img, new_gray, new_edge)
```

```python
butterfly = cv2.imread('./samples/butterfly.jpg', 1)
cube = cv2.imread('./samples/cube.jpg', 1)
```

```python
query_images = [cube, butterfly]
```

```python
for i in range(len(query_images)):
    edge_detector(query_images[i])
    print('')
```
