```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fig_size(d):
    plt.rcParams['figure.figsize'] = [d, d]
```

```python
d = 13
fig_size(d)
```

```python
base = cv2.imread('./samples/base.jpg', 1)
base_rgb = cv2.cvtColor(base, cv2.COLOR_BGR2RGB)
base_gray = cv2.cvtColor(base_rgb, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(base)
```

```python
template = cv2.imread('./samples/template.jpg', 1)
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
```

```python
fig_size(1)
plt.imshow(template)
```

```python
template.shape
```

```python
w, h = template.shape[0], template.shape[1]
```

```python
result = cv2.matchTemplate(base_gray, template_gray, cv2.TM_CCOEFF_NORMED)
threshold = 0.8

loc = np.where(result >= threshold)
```

```python
for pt in zip(*loc[::-1]):
    cv2.rectangle(base_rgb, pt, (pt[0]+w, pt[1]+h), (255,155,250), 2)
```

```python
cv2.imwrite('processed.png', base_rgb)
cv2.imshow('new',base_rgb)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
